# Nürnberger Stadtteile

Plotten von Nürnberger Stadtteilen und markieren von Punkten.

Quelle Shapefiles: https://wiki.openstreetmap.org/wiki/N%C3%BCrnberg/Stadtteile

```python
# Installation von Paketen
# !pip install -q -U pyshp matplotlib
```

```python
import shapefile as shp
import matplotlib.pyplot as plt
import matplotlib.patches as patches

import statistics
```

```python
sf = shp.Reader('Geometrie_NUE_Bezirke/NUE_STAT_BEZIRK.shp', encoding='latin-1')
```

## Liste von Stadtteilen

```python
for shape in sf.shapeRecords():
    print(shape.record[2])
```

## Plotten von Nürnberg bzw. einzelnem Stadtteil inkl. Marker



### Input

Anpassungen können hier gemacht werden:

```python
# 'nürnberg' für komplette karte oder Stadtteil angeben um nur einzelnen Stadtteil zu plotten
# Auch teile von Stadtteilen können angegeben werden z.B. Langwasser plottet dann alle Stadtteile in denen Langwasser vorkommt.
stadtteil = 'nürnberg' 

marker_text = "Frankencenter"
latitude = 49.4036
longitude = 11.1342
```

### Plotting

```python
fig, ax = plt.subplots(figsize=(18,18), frameon=False)
fig.patch.set_visible(False)
ax.axis('off')



for shape in sf.shapeRecords():
    if stadtteil.lower() in shape.record[2].lower() or stadtteil.lower() == 'nürnberg':
        x = [i[0] for i in shape.shape.points]
        y = [i[1] for i in shape.shape.points]
        
        ax.plot(x,y)
        ax.text(statistics.mean(x)-200, statistics.mean(y), shape.record[2])
        
        
def coords(y_coord,x_coord):
    """Konvertierung von GPS-Koordinaten in Shapefile-Koordinaten"""
    x = 71942.42162827333
    x_t = 3636345.1531709

    y = 113348.34040905292
    y_t = -125540.5986282304
    
    return x_coord * x + x_t, y_coord * y + y_t

x, y = coords(latitude, longitude)

plt.axhline(y, color='b', linestyle='-', alpha=0.5)
plt.axvline(x, color='g', linestyle='-', alpha=0.5)

rect = patches.Circle((x, y), 20, linewidth=5, edgecolor='r', facecolor='none')
ax.add_patch(rect)
ax.text(x+15, y+30, marker_text, size='large')


fig.savefig(f'{stadtteil.lower()}.pdf')
fig.savefig(f'{stadtteil.lower()}.svg')
```
